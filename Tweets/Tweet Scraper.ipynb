```python
import snscrape.modules.twitter as sntwitter 
import numpy as np 
import pandas as pd
```

# Scraper 

This is a script that finds popular Tweets with the keyword 'bitcoin' or 'BTC'and then compiles them within a dataset.

## Methodology

- Want to scrape tweets from that day, so we need to establish some parameters that give us "good" tweets. (IE the tweets that we scrape provide better insight) 

    - We set a minimum bound for the number of retweets and favorites: 200
    - This number could be changed, based on further research

- We set want to set up date ranges via formatting. 
- What we can do is for every individual day, we can scrape 300 tweets 




```python
#parameters

#words that we will combine
keyword = ['Bitcoin', '$btc', 'Ethereum', '$eth', 'Solana','$sol', 'Cardano', '$ada', 'Ripple', '$xrp', 'Polkadot', '$dot', 'Avalanche', '$avax', 'Dogecoin', '$doge', 'Shiba Inu', '$shib', 'Polygon', '$matic']

#i, i + 1
date_list = pd.date_range(start="2022-01-01",end="2022-06-01", freq="D")
date_list = date_list.values.astype(str)

dt = []

for i in date_list:
    i = i[:10]
    dt.append(i)

#get popular tweets 
min_retweets = 200
min_favorites = 200
```

```python
def scraper_function(word, dates, rt, fav, maxtw):
    
    tweets_list = []
    
    #between the specified dates: 
    for j in range(len(dates)-1): 
        
        for w in word:

            #get the specific tweet from search 
            for i,tweet in enumerate(sntwitter.TwitterSearchScraper(w + ' since:{} until:{} lang:"en" min_faves:{} min_retweets:{} '.format(dates[j], dates[j+1], fav, rt)).get_items()):

                if i > maxtw:
                    break

                #columns collected 
                tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.location,
                                         tweet.likeCount,tweet.retweetCount, tweet.user.followersCount])
                
                #sanity check
                print("tweet for topic {}: {}/{} scraped. Appending now.".format(w, i, maxtw))
    
    #create a dataframe with values and columns 
    df = pd.DataFrame(tweets_list, columns = ['Date', 'ID', 'Tweet', 'Username', 'Location', 'Favorites', 'Retweets', 'Followers'])

    return df

```

```python
df = scraper_function(keyword, dt, 100, 100, 200)
```

```python
df.to_csv('Crypto_Tweets.csv', index = False)
```
