```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import datetime
```

```python
df = pd.read_csv('data/Full_combined_tickers.csv')
df = df.reset_index(drop=True)
df = df.drop(df.columns[[0, 1]], axis=1) 
df.head()
```

```python
print(df.dtypes)
```

```python
print(df['symbol'].value_counts()) #391 products in total
```

```python
cur_usd = np.unique(df["symbol"])
print(len(cur_usd))
```

```python
df_usd = df.loc[df["symbol"] == cur_usd[0]]
df_usd = df_usd.sort_values(by='datetime')
df_usd["datetime"] = pd.to_datetime(df_usd["datetime"])
df_usd.set_index("datetime", inplace=True)
df_usd = df_usd.drop(['symbol'], axis=1)
df_usd.columns = [cur_usd[0]+'_open', cur_usd[0]+'_high', cur_usd[0]+'_low', cur_usd[0]+'_close', cur_usd[0]+'_volume']

for i in range(1, len(cur_usd)):
    df_temp = df.loc[df["symbol"] == cur_usd[i]]
    df_temp = df_temp.sort_values(by='datetime')
    df_temp["datetime"] = pd.to_datetime(df_temp["datetime"])
    df_temp.set_index("datetime", inplace=True)
    df_temp = df_temp.drop(['symbol'], axis=1)
    df_temp.columns = [cur_usd[i]+'_open', cur_usd[i]+'_high', cur_usd[i]+'_low', cur_usd[i]+'_close', cur_usd[i]+'_volume']
    
    df_usd = df_usd.join(df_temp)

df_usd
```

```python
df_usd.to_csv("data/flattened_data.csv")
```
