```python
df2 = pd.read_csv("data/binance_data.csv")
```

```python
df2
```

```python
df2['datetime'] = df2['datetime']- 18000000
df2['datetime'] = pd.to_datetime(df2['datetime'], unit = 'ms')
df2.head()
```

```python
df2
```

```python
# change datetime to string format
df2['datetime'] = df2['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df2 = df2.drop(df2.columns[[0]], axis=1) 
df2
```

```python
df.columns = ['datetime', 'low', 'high', 'open', 'close', 'volume', 'symbol'] # changeing the column names
```

```python
df = df[['datetime', 'open', 'high', 'low', 'close', 'volume', 'symbol']] # reordering the columns
```

```python
combined_df = pd.concat([df, df2]) # combining the two dataframes
combined_df.shape
```

```python
# removing duplicates from the combined dataframe and sorting it by datetime and symbol
# for duplicate keeping coinbase data is more reasonable choice since we are gonna use S&P500 and we are in US 
# there were 76567 duplicate rows 
combined_df = combined_df.drop_duplicates(subset=['datetime', 'symbol'], keep='first')
```

```python
combined_df.to_csv("data/combined_crypto.csv") # saving the dataframe to a csv file
```
